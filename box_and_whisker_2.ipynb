```python
# Handle the data from the CSV
import csv

csv_data = []
with open('data/combined_tweets.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            print(f'Column names are {", ".join(row)}')
            line_count += 1
        else:
            csv_data.append(row)
            line_count += 1
```

```python
# Remove zero-retweet tweets
def screendata(data, threshold = 1) :
    ret = []
    for vizornoviz in data :
        ret.append([])
        for dat in vizornoviz :
            if dat >= threshold :
                ret[-1].append(dat)
    return ret
```

```python
def buildBins(min = 0, spacing = 1, numbins = 10) :
    ret = [min + float(x)*spacing for x in range(numbins)]
    return ret

def binCenters(bins):
    mids = []
    for i in range(len(bins)-1) :
        lower = bins[i]
        upper = bins[i+1]
        mid = (lower + upper) / 2
        mids.append(mid)
    return mids

```

```python
# Get all of the tweets for all users and split between noviz and viz
def splitdata(csv_data) :
    viz = []
    noviz = []
    for tweet in csv_data :
        if tweet[-1] == "True":
            viz.append(int(tweet[6]))
        else:
            noviz.append(int(tweet[6]))
    return [noviz, viz]
#### Returns noviz first
```

```python
def baw(checkname, screen = False, screenvalue = 1) :
    print("---")
    print(checkname)
    sgmd = []
    for line in csv_data:
        name = line[1]
        if name == checkname :
            sgmd.append(line)

    splitSGMD = splitdata(sgmd)
    
    if screen == True :
        splitSGMD = screendata(splitSGMD, screenvalue)

    # Creating plot
    print("noviz n = " + str(len(splitSGMD[0])))
    print("viz n = " + str(len(splitSGMD[1])))
    plt.boxplot(splitSGMD, notch=False, showfliers=False, labels=["no-viz (n=" + str(len(splitSGMD[0])) + ")", "viz(n=" + str(len(splitSGMD[1])) + ")"])
    plt.ylabel("Retweets")
    plt.title("Response to Tweets by @"+checkname)

    # show plot
    plt.show()
```

```python
def pmorethan(dataset, cutoff) :
    counter = 0
    for dat in dataset :
        if dat > cutoff :
            counter += 1
    return counter / len(dataset)
```

```python
userlib = {}
for line in csv_data:
    user = line[1]
    retweets = int(line[4])
    viz = line[-1]
    if user not in userlib:
        userlib[user] = [[],[]]
    if viz == "True" :
        userlib[user][1].append(retweets)
    else:
        userlib[user][0].append(retweets)
        
print(userlib)

        
```

```python
def mean(l) :
    counter = 0.0
    for element in l :
        counter += element
    if len(l) == 0:
        return None
    return (counter / len(l))

outlib = {}
for user in userlib:
    val = userlib[user]
    print(val[0])
    if mean(val[1]) != None and mean(val[0]) != None :
        outlib[user] = mean(val[1])-mean(val[0])
```
