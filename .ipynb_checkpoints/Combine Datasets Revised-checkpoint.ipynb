```python
import sys, os
import pandas as pd
import numpy as np
```

```python
#read csv files and create an new combined dataframe.
files = os.listdir("data")
dfs  = []
for file in files:
    if file =="combined_tweets.csv":
        continue;

    if os.stat("data/"+file).st_size < 10:
        print(file)
        continue
        
    df = pd.read_csv("data/"+file)
    if len(df)==0:
        continue
    dfs.append(df)
pd.concat(dfs)[['tweet_id', 'user', 'text', 'date',
       'retweets', 'replies', 'likes', 'quotes', 'has_viz']].to_csv("data/combined_tweets.csv",index=False)
```

```python
tweets = pd.read_csv("data/combined_tweets.csv")
tweets
```

```python
remove = ["DNC","BillGates","nytimes"] 
user_counts = pd.DataFrame(tweets["user"].value_counts())
user_counts =user_counts.rename(columns={"user": "num_tweets"})
user_counts = user_counts[user_counts["num_tweets"]>150]
new_tweets = tweets.join(user_counts,how="inner",on="user")
new_tweets
```

```python
new_tweets.to_csv("data/combined_tweets_filtered.csv",index=False)
```
