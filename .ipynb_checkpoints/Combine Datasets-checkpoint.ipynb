```python
import sys, os
import pandas as pd
```

```python
#read csv files and create an new combined dataframe.
files = os.listdir("data")
dfs  = []
for file in files:
    if file =="combined_tweets.csv":
        continue;
    dfs.append(pd.read_csv("data/"+file))
```

```python
pd.concat(dfs).to_csv("combined_tweets.csv")
```
