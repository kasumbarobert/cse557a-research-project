```python
import sys, os
import pandas as pd
```

```python
#read csv files and create an new combined dataframe.
files = os.listdir("data")
dfs  = []
for file in files:
    if file =="combined_tweets.csv":
        continue;
    df = pd.read_csv("data/"+file)
    if len(df)==0:
        continue
    dfs.append(df)
pd.concat(dfs)[['tweet_id', 'user', 'text', 'date',
       'retweets', 'replies', 'likes', 'quotes', 'has_viz']].to_csv("data/combined_tweets.csv",index=False)
```
